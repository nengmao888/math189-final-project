## Section 1: Introduction

Housing affordability is a complex crisis in San Diego, especially for students at UC San Diego residing in high-cost areas like La Jolla. While rent price is the most immediate metric of affordability, it is often linked to environmental quality, specifically noise pollution and urban infrastructure. This project aims to investigate the socioeconomic and environmental tradeoffs inherent in San Diego’s housing market. We seek to understand how various urban factors, such as tree canopy coverage and transit ridership, correlate with noise levels, and how these factors collectively influence housing value scores and demographic distributions. By framing noise pollution not just as a nuisance but as a public health hazard and an economic indicator, this study will explore whether disadvantaged neighborhoods bear a disproportionate burden when it comes to noise, and whether affordable housing often comes at the cost of long-term health outcomes associated with that exposure.

## Section 3: Data Cleaning and Merging
A core challenge of this project was that our datasets came from different geographic units. Noise data was available at the **census tract** level, while demographic and housing data used **ZCTAs** and **ZIP codes**. Combining them required a careful spatial merging process.
### 3.1 The Geographic Mismatch Problem
Census tracts, ZCTAs, and ZIP codes are three different ways of dividing up space, and they do not line up neatly. A single ZCTA may overlap with several census tracts, and a single census tract may span multiple ZCTAs. To move from tract-level noise data to ZCTA-level noise estimates, we needed a principled way to aggregate across these boundaries.
### 3.2 Tract-to-ZCTA Aggregation
We used the HUD TRACT-to-ZIP crosswalk (December 2022) to map each California census tract to its corresponding ZCTA.
### 3.3 Merging All Datasets
Once noise data was aggregated to the ZCTA level, we merged all five datasets on the `zcta` key. The Zillow housing data required an additional step: Zillow uses ZIP codes, not ZCTAs, so we aligned them using the same HUD crosswalk. The final merged dataset (`master_dataset.csv`) contains 154 rows (one per ZCTA) and 17 columns.
### 3.4 Missing Data
The master dataset has substantial missing data, primarily in the demographic and socioeconomic variables. This is a known issue with ACS data at the ZCTA level — smaller or less-populated ZCTAs are often suppressed in Census releases to protect respondent privacy.
