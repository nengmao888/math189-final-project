## 1. Transportation Noise Data

**Source:** U.S. DOT Bureau of Transportation Statistics — CONUS Road Noise 2020 (raster), aggregated to San Diego County census tracts via zonal statistics.

**Geography:** Census tract (San Diego County)

**Variables:**
| Column | Description |
|--------|-------------|
| `tract_id` | Census tract GEOID (string) |
| `tract_name` | Tract name |
| `noise_mean_db` | Mean transportation noise level (dB) |
| `noise_max_db` | Max noise level (dB) |
| `noise_min_db` | Min noise level (dB) |

```python
# Load transportation noise data
noise = pd.read_csv('../data/processed/tract/tract_noise.csv')

print(f'Shape: {noise.shape}')
print(f'\nMissing values:')
print(noise.isna().sum())
print(f'\nData types:')
print(noise.dtypes)
```

## 3. ZIP-to-ZCTA Crosswalk

**Source:** UDS Mapper ZIP-to-ZCTA crosswalk.

**Purpose:** ZIP codes (used by Zillow) do not map 1-to-1 with ZCTAs (used by Census). This crosswalk links the two geographies.

**Variables:**
| Column | Description |
|--------|-------------|
| `zip_code` | USPS ZIP code (string) |
| `zcta` | Corresponding ZCTA code (string) |

```python
crosswalk = pd.read_csv('../data/processed/zcta/zcta_crosswalk.csv')

print(f'Shape: {crosswalk.shape}')
print(f'\nMissing values:')
print(crosswalk.isna().sum())
print(f'\nData types:')
print(crosswalk.dtypes)
```

```python
crosswalk.head()
```
