## Load Transportation Noise Data

This dataset contains transportation-related noise exposure measurements by census tract.

```python
# Load transportation noise data

noise = pd.read_csv(
    '../data/processed/tract/sd_tract_noise.csv'
)

# Preview first rows
noise.head()
```

```python
# Check dataset dimensions

print(noise.shape)
```

```python
# Check column names

print(noise.columns)
```

```python
# Check data types

noise.dtypes
```

```python
# Check missing values

noise.isna().sum()
```

```python
# Standardize tract identifiers

noise['GEOID'] = noise['GEOID'].astype(str)
```

```python
# Rename columns for consistency

noise = noise.rename(columns={
    'GEOID': 'tract_id',
    'NAME': 'tract_name',
    'mean_noise': 'noise_mean_db',
    'max_noise': 'noise_max_db',
    'min_noise': 'noise_min_db'
})

noise.head()
```

```python
# Save cleaned transportation noise dataset

noise.to_csv(
    '../data/processed/tract/tract_noise.csv',
    index=False
)
```
