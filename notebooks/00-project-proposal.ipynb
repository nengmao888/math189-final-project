# Project Proposal

## 1. Research Question

What factors influence product ratings?

In this project, we aim to investigate how different features of products (such as price, category, number of reviews, and other attributes) affect their ratings.

---

## 2. Motivation

Understanding the factors that influence product ratings is important for both consumers and businesses. 

For consumers, it helps them make better purchasing decisions.  
For businesses, it provides insights into how to improve product quality and customer satisfaction.

---

## 3. Data Sources

We plan to use publicly available datasets, such as:

- Amazon product reviews dataset  
- Kaggle datasets on product ratings  
- Other open-source e-commerce datasets  

---

## 4. Data Collection

The data will be obtained from online sources such as Kaggle or publicly available datasets.

We will download and clean the dataset, focusing on relevant variables such as ratings, price, and number of reviews.

---

## 5. Proposed Analysis

We plan to perform the following analyses:

- Exploratory Data Analysis (EDA)  
- Correlation analysis between features and ratings  
- Regression models to predict ratings  
- Statistical testing to evaluate significance  

---

## 6. Expected Outcomes

We expect to identify key factors that significantly influence product ratings.

Additionally, we aim to build a model that can predict ratings based on product features.

---

## 7. Feasibility

This project is feasible within the course timeline because:

- Data is publicly available  
- Methods are standard statistical techniques  
- The workload is appropriate for a group project
