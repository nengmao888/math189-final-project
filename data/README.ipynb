# Data Folder

This folder contains all datasets used in the project.

- `raw/`: Original data (should not be modified)
- `processed/`: Cleaned or transformed data

All analysis in the notebooks is based on these datasets.

# GitHub Data Policy

Due to GitHub file size limitations, very large GIS raster files and raw ZIP archives are stored locally and are not tracked in the repository.

The repository mainly contains:
- cleaned CSV datasets
- processed tract/ZCTA tables
- merged analysis-ready datasets
- preprocessing notebooks

Team members should work primarily with files inside:

processed/

# Raw Data Access

Some raw datasets are excluded from GitHub due to file size limitations.

Project collaborators can access the original raw datasets from the shared Google Drive folder.

Raw datasets include:
- ACS original downloads
- Zillow full datasets
- Transportation noise raster files
- ICPSR demographic files

Please place raw datasets into:

data/raw/

before running the full preprocessing pipeline in:

- math_189_datasets.ipynb

# Data Folder Structure

This folder contains all datasets used for the MATH 189 group project:

## “The Sound of Inequality”

The project investigates relationships between transportation noise exposure and socioeconomic inequality in California, with a focus on Southern California and San Diego County.

# Folder Organization

## raw/

Original downloaded datasets.

Contains:
- ZIP archives
- Raster files
- Original ACS downloads
- Raw Census datasets
- Large GIS files

These files should NOT be modified directly.

Examples:
- CONUS_road_noise_2020.zip
- CONUS_road_and_aviation_noise_2020.zip
- ICPSR_38528-V6.zip

## processed/

Cleaned and analysis-ready datasets.

These datasets were standardized and cleaned for:
- statistical analysis
- regression modeling
- visualization
- spatial joins
- dataset merging

### tract/

Processed Census tract-level datasets.

Examples:
- tract_noise.csv

Variables include:
- mean transportation noise
- max transportation noise
- min transportation noise

### zcta/

Processed ZIP/ZCTA-level datasets.

Examples:
- zcta_housing.csv
- zcta_crosswalk.csv
- zcta_demographics_selected.csv

Variables include:
- home value
- median household income
- poverty rate
- unemployment rate
- demographic indicators

### merged/

Final merged datasets used for statistical modeling and analysis.

This folder will contain:
- merged tract/ZCTA datasets
- analysis-ready regression tables
- final project datasets

# Data Cleaning Workflow

Raw datasets were processed in:

- math_189_datasets.ipynb

The notebook includes:
- data loading
- variable selection
- missing value handling
- ZIP/ZCTA standardization
- demographic extraction
- cleaned CSV export

Analysis workflows and spatial merges are performed in:

- cleaned_datasets.ipynb

# Main Geographic Units

The project primarily uses:
- Census Tracts
- ZCTAs (ZIP Code Tabulation Areas)

ZIP-to-ZCTA crosswalk tables are used to align datasets from different geographic systems.

# GitHub Data Policy

Due to GitHub file size limitations:

- large raster files
- GIS layers
- ZIP archives
- raw ACS downloads

may be stored locally and excluded from repository tracking.

Processed CSV files are preferred for:
- collaboration
- reproducibility
- version control
