# MATH 189 Course Project

---

## Grading Breakdown

Your course project score will be aggregated as follows:

- 20% Proposal  
- 25% Video Presentation  
- 35% Final Report  
- 20% Peer Evaluations  

---

# Proposal

Your project should be a **novel proposal for a statistical application**. You will be graded for originality.

For starters, your project should **not be a direct adaptation of an existing project found online**, e.g., a Kaggle project analysis that someone else has already done.

Your proposal needs to be submitted as a **PDF file on Gradescope**, and should be **no longer than 2 pages**.

---

## Proposal Guidelines

While formulating your project proposal, focus on the following:

- Why is this problem important?  
- What are the data sources for this?  
- Where do you plan on acquiring the data from?  
- What type of analyses do you propose for the dataset?  
- What are your expected outcomes for the project?  

---

## Proposal Grading Rubric

### Clarity (40%)
Is the proposal clear, well-organized, and easy to understand?  
Are the objectives and goals precisely defined?

### Difficulty and Originality (40%)
You will lose points if either:
- (i) the project is too simple and lacks originality, or  
- (ii) does not provide a sufficient challenge for a ~3 person project.

### Feasibility (20%)
The proposal includes a realistic plan and is feasible within the course timeline.

---

# Final Report

Your final report should be a **comprehensive document** that details your project from start to finish.

You should describe all the steps you took in a **clear and concise manner** and include:

- Tables  
- Plots  
- Analyses  
- Visualizations  

Your report should resemble an **English-style report supported by scientific evidence and statistical analysis**.

---

## Final Report Checklist

- A statement of the problem you are investigating  
- Why is this problem relevant? What inspired you?  
- Where did you get the data from?  
- Description of the data  
- Exploratory data analysis (EDA)  
- Existing analyses on this dataset (with references)  
- Statistical methods used  
- Interpretation of results  
- Limitations and shortcomings  

---

⚠️ **Important Note**

Not all projects need to end with a conclusive result.

It is acceptable to conclude:
- "We need more data"  
- "We need further analysis"  

However, you must **justify your reasoning using statistical thinking**.

---

## Final Report Grading Rubric

### Clarity
The report must be clear, concise, and well-organized.

### Exploratory Data Analysis
- Have you understood the dataset?  
- Identified missing values, outliers, or nuisance variables?  
- Explained how they were handled?  

### Statistical Analyses
- Are the methods appropriate?  
- Are assumptions clearly stated and justified?  

### Interpretation of Results
- Are conclusions clear and consistent with analysis?  
- Are results easy to understand?  

### Relation to Proposal
- Does the project align with the original proposal?  
- If not, is the deviation justified?  

### Code Execution
- Code must be well-documented  
- Code must be reproducible  
- TAs should be able to run it without issues

# Video Presentation

In addition to the written report, your group must submit a **recorded video presentation**.

This is your opportunity to **present and "sell" your analysis**.

---

## Requirements

- Format: Video file (.mp4 or .mov) uploaded to Canvas  
- Length: Less than 20 minutes  
- Tools: Zoom, OBS, Loom, or similar  
- Optional: Slides or RISE notebook slides  

---

## Suggested Presentation Outline

- Problem statement and motivation  
- Data sources and data cleaning  
- Exploratory Data Analysis (EDA)  
- Statistical methods and assumptions  
- Results and interpretation  
- Limitations and future work  

---

## Presentation Grading Rubric

### Statistical Communication (60%)
- Accurate explanation of findings  
- Correct interpretation (coefficients, p-values, etc.)  
- Understanding of real-world implications  

### Presentation of Results (40%)
- Clear and cohesive narrative  
- Focus on key insights  
- Avoid simply reading slides
