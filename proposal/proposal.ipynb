## Introduction

Housing affordability is a complex crisis in San Diego, especially for students at UC San Diego residing in high-cost areas like La Jolla. While rent price is the most immediate metric of affordability, it is often linked to environmental quality, specifically noise pollution and urban infrastructure. This project aims to investigate the socioeconomic and environmental tradeoffs inherent in San Diego’s housing market. We seek to understand how various urban factors, such as tree canopy coverage and transit ridership, correlate with noise levels, and how these factors collectively influence housing value scores and demographic distributions. By framing noise pollution not just as a nuisance but as a public health hazard and an economic indicator, this study will explore whether disadvantaged neighborhoods bear a disproportionate burden when it comes to noise, and whether affordable housing often comes at the cost of long-term health outcomes associated with that exposure.


## Data

To conduct a statistical analysis, we will be using datasets from multiple sources, both conventional and unconventional. Demographic and socioeconomic metrics, including median household income, ethnic makeup, and population density, will be sourced from the U.S. Census Bureau and San Diego Open Data portals. For environmental variables, we will utilize tree canopy land cover classifications provided by SANDAG (San Diego Association of Governments) and transit ridership patterns. The core of our noise analysis will rely on the National Transportation Noise Map and the SAN QHP Dashboard, which provides granular data on aircraft and traffic noise. Finally, to bridge these factors with housing, we will collect rental and pricing data from Zillow Research and simplemaps. This is not a cumulative list and we expect to incorporate more variables (median home age, average cost of insurance, even untested and unconventional factors like distance from university) if data availability permits.
For data integration we use Spatial Joins and GeoJSON normalization. The primary challenge involves the spatial alignment of these different datasets like merging ZIP codes, Census Tracts, and GeoJSON shapes. This requires data cleaning and normalization on a large scale  to ensure a consistent unit of analysis. Once data cleaning is complete, we can move on to an exploratory data analysis (EDA) that will visualize and give us an initial understanding of the data on hand.


## Methods

Following our data portion, our approach will move from exploratory correlation to formal hypothesis testing and predictive modeling. We’ll start by exploring correlation between noise and various different variables by plotting them. Then, we propose building a Multiple (Multivariate) Linear Regression model to quantify the relationship between environmental factors and noise levels. An example, with some of the key factors that we expect, is as follows:


Example regression model:

$$
\text{Noise Exposure} \sim \text{Median Income} + \text{Tree Canopy} + \text{Transit Activity} + \text{Population Density}
$$
In order to make sure the model is accurate, we will check Variance Inflation Factors (VIF) for multicollinearity and use Q-Q plots to ensure the normality of residuals. Furthermore, we can use the permutation test and KS test to compare the distribution between high noise and low noise clusters. 

## Expected Outcomes

An example of a null hypothesis that we can test for is: there is no statistically correlation between a neighborhood’s tree canopy coverage and its average ambient noise level. Conversely, our alternative hypothesis would be: increased vegetation serves as a significant biological buffer, reducing noise pollution. We can do this for all kinds of different variables. 

For some of our results, we expect to identify a clear inverse correlation between urban tree canopy coverage and noise levels, as well as a positive correlation between high-noise zones and lower socioeconomic status (SES), worse health outcomes for residents in high noise zones, and many more. We hypothesize that our analysis will likely reveal that more "affordable" areas for students and low-income residents are frequently situated in "pollution hotspots" or regions with high transit-related noise (near highways or the trolley). We hope to produce a comprehensive visualization of these disparities, highlighting neighborhoods where the tradeoff between rent and environmental quality is most extreme. Ultimately, these findings could provide a data-driven basis for proposing urban planning interventions, such as targeted soundproofing or reforestation in disadvantaged neighborhoods, to improve both housing equity and public health outcomes in San Diego.
